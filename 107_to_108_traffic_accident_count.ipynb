```python
import pandas as pd
import numpy as np
```

```python
# 讀檔

df107_A1 = pd.read_csv('D:/Python/csv/traffic/107A1.csv',encoding='utf-8')
df107_A1.dropna(axis=0, how='any', inplace=True)#刪除空欄位

df107_A2 = pd.read_csv('D:/Python/csv/traffic/107A2.csv',encoding='utf-8')
df107_A2.dropna(axis=0, how='any', inplace=True)#刪除空欄位

df108_A1 = pd.read_csv('D:/Python/csv/traffic/108A1.csv',encoding='utf-8')
df108_A1.dropna(axis=0, how='any', inplace=True)#刪除空欄位

df108_A2 = pd.read_csv('D:/Python/csv/traffic/108A2.csv',encoding='utf-8')
df108_A2.dropna(axis=0, how='any', inplace=True)#刪除空欄位
```

```python
# 將107年的A1跟A2Merge
df107 = df107_A1.append(df107_A2, ignore_index=True)
df107
```

```python
# 將108年的A1跟A2Merge
df108 = df108_A1.append(df108_A2, ignore_index=True)
df108
```

```python
#切割車種內的車子類型
car_type107 = df107['車種'].str.split(r';',expand=True)
car_type108 = df108['車種'].str.split(r';',expand=True)
```

```python
#合併所有欄位成一行
all_car_type107 = car_type107[0]
for i in range(1,24):
       all_car_type107 = pd.concat([all_car_type107,car_type107[i]],axis=0,ignore_index=True) 
all_car_type108 = car_type108[0]
for i in range(1,24):
       all_car_type108 = pd.concat([all_car_type108,car_type108[i]],axis=0,ignore_index=True) 
```

```python
type_count107 = all_car_type107.to_frame()[0].value_counts()
type_count107.to_frame()
```

```python
type_count108 = all_car_type108.to_frame()[0].value_counts()
type_count108.to_frame()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(20,10))
type_count107.head(10).plot(kind='bar')
plt.title('107年前十傷亡總數長條圖',fontsize=20)
plt.xticks(rotation=45)
plt.xticks(fontsize=15)

def createLabels(data):
    i=0
    for item in data:
        height = item
        plt.text(
            i, 
            height+15, 
            '%d' % int(height),
            ha = "center",
            va = "bottom",
        )
        i+=1
createLabels(type_count107.head(10))
plt.tight_layout()
plt.savefig('107_top10.png',pad_inches=50.0)
plt.show()
```
